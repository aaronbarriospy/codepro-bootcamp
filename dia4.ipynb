Diccionarios
Un diccionario es una coleccion de pares formados por una clave y un valor asociado a la clave. Se construyen poniendo los pares entre "{}" separados por comas y separando la clave del valor con dos puntos.

Se caracterizan por:

-No tienen orden
-Pueden contener elementos de distintos tipos
-Son mutables, es decir, pueden alterarse durante la ejecucion de un programa
-Las claves son unicas, es decir, no pueden repetirse en un mismo diccionario
-Los valores de un diccionario pueden ser de cualquier tipo de datos: int, float, bool, list, dict, etc.
-Son iterables

```python
class Coche:
    tipo = 'vehiculo'

    # Atributos de instancia
    def __init__(self, modelo, posicion = 0):
        self.modelo = modelo
        self.posicion = posicion

    # Definir otras acciones de la clase
    def acelerar(self, metros):
        self.posicion += metros
        print(f'EL vehiculo modelo {self.modelo} acelera y avanza {metros}. Nueva posicion {self.posicion}')

    def girar(self, direccion):
        print(f'El {self.modelo} avanza y gira a la {direccion}')

coche1 = Coche('Toyota', 5)
coche1.acelerar(1)
```

Herencia

No es una erencia como tal, sino una relacion entre clases.


```python
class AnimalDeGranja:
    # atributos de clase
    tipo = 'animal de granja'

    # atributos de instancia
    def __int__(self, nombre, color, sonido):
        self.nombre = nombre
        self.color = color
        self.sonido = sonido

    # Metodos (acciones de la clase)
    def hablar(self):
        print(f'Hablo y digo {self.sonido}')

    def presentarse(self):
        print(f'Soy {self.nombre}')


class Gallina(AnimalDeGranja):
    huevos_por_dia = 1
    def poner_huevo(self):
        print(f'Estoy poniendo {self.huevos_por_dia} huevos por dia')



class Vaca(Gallina):
    patas = 4

    def mujir(self):
        print("Muuuu")


gallina1 = Gallina('')
gallina1.hablar()
gallina1.presentarse()

vaca = Vaca('Lola', 'blanco y negro', 'Muuu')
vaca.hablar()
```

Observaci√≥n: Editar las clases hijas para que no se inicialicen

```python
class Autos:
    tipo = 'rodado'

    def __init__(self, marca, ruedas, puertas):
        self.marca = marca
        self.ruedas = 4
        self.puertas = puertas
        

class Lamborghini(Autos):
    def rugir(self):
        print(f'El motor del {self.marca} ruge fuerte')
        

class Porsche(Lamborghini):
    def cero_a_cien(self):
        segundos = 6
        print(f'Hace de 0 a 100 en {segundos} segundos')

auto1 = Porsche("Porsche", 4, 2)
auto1.rugir()
auto1.cero_a_cien()
auto2 = Lamborghini('Lamborghini', 4, 2)
auto2.rugir()
```
