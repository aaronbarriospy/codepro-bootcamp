# Bucle

While es una palabra reservada que nos permite ejecutar una serie de instrucciones mientras una condicion se cumpla. Al momento que la condicion deje de ser cierta el bucle va a parar


```python
numero_a_adivinar = 20
numero_ingresado = 10

# While + Condicion
while numero_ingresado != numero_a_adivinar:
    numero_ingresado = int(input('Ingresa un numero'))
    print(numero_ingresado)
```

# Bucles infinitos

Un bucle infinito sucede cuando la condicion siempre es cierta y por lo tanto el bucle no termina de ejecutarse. Break es una palabra reserbada que nos permite salir de un bucle.

```python
numero = 30

while True:
    numero_a_ingresar = int(input('ingresa un numero'))
    if numero_a_ingresar == numero:
        print('El nro es correcto')
        break
    else:
        print('El nro es incorrecto, intentalo de vuelta')
```

```python
while True:
    print("Hola")
    break
## Si no sacamos el break, va a haber un bucle infinito ##
```
