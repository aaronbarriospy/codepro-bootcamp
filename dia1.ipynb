# Reglas para nombrar variables
Algunas reglas y buenas practicas para nobras variables en Python:
-Puede incluir letras, numeros y guiones bajos (_)
-Debe empezar con una letra o guion bajo, nunca con un numero
-Evitar usar palabras reservadas de Python (como print, def, if, etc.)
-Usar nombres descriptivos (no pongas x, a, b si no hace falta!). En programacion un espacio vacio es un caracter, por esop al escribir variables no deben existir espacios vacios
